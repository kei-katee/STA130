````python
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd
import numpy as np
from plotly.subplots import make_subplots
from palmerpenguins import load_penguins

# Load the Palmer Penguins dataset
df = load_penguins()

# Remove any rows with NaN in flipper_length_mm or species columns
df = df.dropna(subset=['flipper_length_mm', 'species'])

# List of species
species_list = df['species'].unique()

# Create a subplot for histograms of each species
fig = make_subplots(rows=1, cols=len(species_list), subplot_titles=species_list)

# Loop over each species
for i, species in enumerate(species_list):
    # Filter data for the species
    species_data = df[df['species'] == species]['flipper_length_mm']
    
    # Calculate mean, median, range, IQR, and standard deviation
    mean = species_data.mean()
    median = species_data.median()
    min_val = species_data.min()
    max_val = species_data.max()
    std_dev = species_data.std()
    q1 = species_data.quantile(0.25)
    q3 = species_data.quantile(0.75)
    
    # Calculate 2 standard deviations from the mean
    lower_2std = mean - 2 * std_dev
    upper_2std = mean + 2 * std_dev

    # Add histogram for the species
    fig.add_trace(
        go.Histogram(x=species_data, name=species, opacity=0.75),
        row=1, col=i+1
    )

    # Add mean and median lines
    fig.add_vline(x=mean, line=dict(color='blue', dash='dash'), row=1, col=i+1)
    fig.add_vline(x=median, line=dict(color='green', dash='dot'), row=1, col=i+1)

    # Add range (min to max) rectangle
    fig.add_vrect(x0=min_val, x1=max_val, fillcolor='lightblue', opacity=0.2, line_width=0, row=1, col=i+1)
    
    # Add IQR rectangle
    fig.add_vrect(x0=q1, x1=q3, fillcolor='lightgreen', opacity=0.3, line_width=0, row=1, col=i+1)
    
    # Add 2 standard deviations rectangle
    fig.add_vrect(x0=lower_2std, x1=upper_2std, fillcolor='orange', opacity=0.2, line_width=0, row=1, col=i+1)

# Update layout for clarity
fig.update_layout(title="Histograms of Flipper Length by Species with Statistical Annotations",
                  barmode='overlay', showlegend=False, height=400, width=1000)

# Show the plot
fig.show()
````

````python
import seaborn as sns
import plotly.graph_objects as go
import numpy as np
from scipy.stats import gaussian_kde

# Load the penguins dataset
penguins = sns.load_dataset("penguins")

# Function to create Plotly KDE plot with location and scale markers
def plot_kde_with_statistics(df, species_name):
    flipper_length = df['flipper_length_mm'].dropna()

    # Kernel density estimation using Gaussian KDE (direct computation)
    kde = gaussian_kde(flipper_length)
    x_vals = np.linspace(flipper_length.min(), flipper_length.max(), 500)
    kde_y_vals = kde(x_vals)
    
    # Create Plotly figure
    fig = go.Figure()

    # Plot KDE as a smooth curve
    fig.add_trace(go.Scatter(x=x_vals, y=kde_y_vals, mode='lines', fill='tozeroy', name=f'KDE of {species_name}'))

    # Calculate statistics
    mean = flipper_length.mean()
    median = flipper_length.median()
    std_dev = flipper_length.std()
    q1 = flipper_length.quantile(0.25)
    q3 = flipper_length.quantile(0.75)
    iqr = q3 - q1
    min_val = flipper_length.min()
    max_val = flipper_length.max()

    # Add vertical lines for mean and median
    fig.add_vline(x=mean, line=dict(color='blue', dash='dash'), annotation_text='Mean', annotation_position="top left")
    fig.add_vline(x=median, line=dict(color='green', dash='dash'), annotation_text='Median', annotation_position="top right")

    # Add rectangles for range, IQR, and 2 std devs
    fig.add_vrect(x0=min_val, x1=max_val, fillcolor="gray", opacity=0.1, line_width=0, annotation_text="Range", annotation_position="bottom left")
    fig.add_vrect(x0=q1, x1=q3, fillcolor="orange", opacity=0.2, line_width=0, annotation_text="IQR", annotation_position="bottom left")
    fig.add_vrect(x0=mean - 2 * std_dev, x1=mean + 2 * std_dev, fillcolor="blue", opacity=0.1, line_width=0, annotation_text="2 Std Devs", annotation_position="bottom right")

    # Update layout
    fig.update_layout(
        title_text=f"Flipper Length KDE and Statistics for {species_name}",
        xaxis_title="Flipper Length (mm)",
        yaxis_title="Density"
    )

    # Show the figure
    fig.show()

# Loop through each species and create the KDE plot with statistics
species_list = penguins['species'].unique()
for species in species_list:
    species_df = penguins[penguins['species'] == species]
    plot_kde_with_statistics(species_df, species)
````

1. Histograms
A histogram divides the data into bins and displays the frequency (or density) of values that fall within each bin.

Pros:
Easy to interpret: Histograms provide a straightforward visual representation of the data's shape (modality, skewness, etc.).
Good for large data: Works well with large datasets where binning gives a clear sense of distribution.
Can show the data's modality: Whether the data is unimodal (one peak), bimodal (two peaks), or multimodal (multiple peaks), histograms clearly reflect it.
Customizable bin size: The bin size can be adjusted to focus on different levels of detail in the data distribution.
Cons:
Dependent on bin size: The appearance of the histogram can change dramatically based on the choice of bin size. If the bin size is too large, important details in the data may be lost; if too small, the visualization may become overly "noisy."
Lack of smoothness: Unlike KDEs, histograms can appear blocky or jagged, especially in small datasets or when bin size is not optimal.
Hard to compare distributions: Comparing multiple histograms in a single plot can become cluttered or difficult, especially if the data overlaps significantly.
2. Box-and-Whisker Plots (Boxplots)
A boxplot summarizes the data by its quartiles, highlighting the median, interquartile range (IQR), and potential outliers.

Pros:
Compact representation: Boxplots give a concise summary of the datas spread, including the median, quartiles, and outliers, all in a single plot.
Ideal for comparing multiple groups: Boxplots are especially useful when you need to compare distributions between multiple datasets side by side.
Outlier detection: One of the main strengths of boxplots is their ability to clearly highlight outliers, which might be obscured in histograms or KDEs.
Not sensitive to binning: Unlike histograms, boxplots dont rely on arbitrary choices like bin size.
Cons:
No information on modality: Boxplots provide no information about the shape of the distribution (e.g., whether it's unimodal or bimodal), which histograms and KDEs can show.
Limited detail: Boxplots summarize data using only a few summary statistics (median, IQR, min, max, outliers), so they do not reflect the full structure of the distribution.
No detailed view of distribution tails: While boxplots provide a high-level summary, the tails (outside the whiskers) are generalized, making it difficult to understand their exact shape.
3. Kernel Density Estimators (KDE)
A KDE creates a smooth curve that estimates the probability density function of the data. Its a non-parametric way to estimate the underlying distribution of the data.

Pros:
Smooth representation: KDEs produce a smooth curve, giving a more fluid and accurate visualization of the data's distribution compared to the "blocky" nature of histograms.
Not bin-dependent: Unlike histograms, KDEs are not affected by arbitrary binning choices, which makes them more flexible in representing the data.
Better for small datasets: KDEs can provide more insightful estimates for smaller datasets, where histograms might fail due to limited data points.
Reveals the true distribution: KDEs are particularly useful when you want to detect the underlying distribution (e.g., skewness, multimodal features) more accurately than a boxplot.
Cons:
Bandwidth sensitivity: The appearance of a KDE is highly sensitive to the bandwidth parameter (which controls how smooth or wavy the curve is). A bandwidth thats too high will oversmooth the data, while one thats too low will create a noisy, spiky plot.
Harder to interpret: For non-experts, KDEs might be less intuitive than histograms or boxplots, particularly because they dont show exact data counts or summary statistics directly.
Comparison across multiple groups: Like histograms, comparing multiple KDEs can become difficult due to overlap, though it can still be done (by adjusting transparency or plotting different curves on the same axis).
