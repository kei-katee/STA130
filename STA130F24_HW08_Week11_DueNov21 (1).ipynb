A Classification Decision Tree addresses classification problems, where the goal is to place data points into distinct categories or "classes." The task is essentially to predict a categorical outcome, often a binary result like "Yes/No" or "Spam/Not Spam," but it can extend to multiple classes as well.

Example:
Medical Diagnosis: Predicting if a patient has a specific disease based on symptoms, test results, or other medical information.
Email Filtering: Classifying emails as "Spam" or "Not Spam" based on word frequency, sender details, and other factors.
Customer Segmentation: Classifying customers into categories based on purchasing behavior or demographic characteristics.
Credit Scoring: Predicting whether a person is likely to default on a loan based on financial history, age, and income.

Classification Decision Tree Prediction Process:
A Classification Decision Tree makes predictions by following a sequence of decisions based on specific feature values in the data. Here’s how it works:

Decision Nodes: Starting at the root, the tree evaluates a feature based on a decision rule (e.g., "Is the feature ‘age’ greater than 30?").
Branching: Depending on the answer to the rule, the tree branches out to another node. Each node represents a decision point, based on another feature.
Sequence of Rules: The model moves down the branches of the tree, applying different rules at each level.
Leaf Node (Prediction): Eventually, the path reaches a "leaf node," which provides the final class prediction for the input data (e.g., "Yes" or "No," "Class A" or "Class B"). The outcome is determined by the most common class among the data points in that leaf.
The tree’s structure allows it to capture complex, non-linear relationships in the data, because each path can involve unique, compound rules (combinations of features). As a result, Decision Trees can naturally handle interactions between features by making different decisions depending on the values of preceding features.

Multiple Linear Regression Prediction Process:
In contrast, Multiple Linear Regression predicts outcomes based on a linear equation that models the relationship between the features and the target variable. Here’s the process:

Weighted Sum of Features: The model calculates a prediction by taking each feature, multiplying it by a learned weight (or coefficient), and adding a constant (the intercept). The formula looks like this:

Prediction=𝑤1×𝑥1+𝑤2×𝑥2+⋯+𝑤𝑛×𝑥𝑛+𝑏
where 𝑤1,𝑤2,…,𝑤𝑛 are the feature weights, 𝑥1,𝑥2,…,𝑥𝑛 are feature values, and 
𝑏 is the intercept.

Continuous Outcome: The result is a single continuous value, which is typical for regression problems (e.g., predicting price, temperature, or weight).

Linear Nature: This method doesn’t capture non-linear interactions between features, as it adds feature contributions in a straightforward way. Multiple Linear Regression assumes that each feature contributes independently and additively to the outcome.

1. **Accuracy** measures the proportion of true results (both true positives and true negatives) in the population.

$$\text{Accuracy} = \frac{TP + TN}{TP + TN + FP + FN}$$
    
2. **Sensitivity** measures the proportion of actual positives that are correctly identified.

$$\text{Sensitivity} = \frac{TP}{TP + FN}$$

3. **Specificity** measures the proportion of actual negatives that are correctly identified.

$$\text{Specificity} = \frac{TN}{TN + FP}$$

4. **Precision** measures the proportion of positive identifications that were actually correct.

$$\text{Precision} = \frac{TP}{TP + FP}$$

1 Accuracy is most useful when the classes in the dataset are relatively balanced and when both types of errors (false positives and false negatives) carry similar consequences. When predicting whether it will rain tomorrow, accuracy can be a useful metric since both correct predictions of rain (true positives) and no rain (true negatives) are of value to users. Weather predictions generally have a balanced number of rainy vs. non-rainy days, so accuracy provides a balanced assessment of the model’s performance.

2 Sensitivity is most useful in situations where it’s crucial to identify all actual positive cases, even at the risk of some false positives. In other words, sensitivity is ideal when missing a positive case (false negative) is more costly than mislabeling a negative case (false positive). In a screening test for a serious illness like cancer, sensitivity is key. Detecting every possible case (true positives) is critical because missing a true case (false negative) could have serious consequences for the patient’s health. Even if a few false positives occur, they can be followed up with more precise testing.

3 Specificity is useful in cases where it’s essential to avoid false positives, or when the cost of a false positive is higher than that of a false negative. Specificity helps ensure that negatives are correctly identified, which is important when there are many actual negative cases. In fraud detection, specificity is crucial because marking a legitimate transaction as fraud (false positive) can be highly inconvenient for customers and costly for the bank. Specificity helps ensure that normal transactions are accurately classified, while still catching true instances of fraud.

4 Precision is ideal in scenarios where false positives are costly or when the focus is on ensuring that positive predictions are highly reliable. Precision is especially valuable in situations where you only want to act on a prediction if it’s very likely to be correct.
Precision is important in spam detection because marking legitimate emails as spam (false positives) can cause users to miss important messages. High precision ensures that when an email is flagged as spam, it’s very likely to actually be spam, reducing the chances of false positives.

DecisionTreeClassifier.fit(X, y) trains the Decision Tree Classifier.

The clf2 model is a decision tree classifier trained to predict whether a book is a hard cover or a paper back using the features NumPages, Thick, and List Price. Predictions are made by following a series of binary decisions through the tree:

The process begins at the root node, where the model evaluates a condition on one of the features (e.g., NumPages ≤ 250). Depending on whether the condition is true or false, the observation is routed to the left or right branch of the tree.

At each subsequent node, the model evaluates another condition based on a different feature (e.g., List Price ≤ $20 or Thick ≤ 1.5 inches). This process continues as the model narrows down the possibilities.

Eventually, the observation reaches a leaf node, which represents the final prediction. The class assigned to this leaf node (e.g., hard cover or paper back) is determined by the majority of training samples that fall into that node.

The tree structure ensures that predictions are interpretable, with each split focusing on a feature and threshold that helps separate the classes effectively.

````python
# Prepare the test data
y_test = pd.get_dummies(ab_reduced_noNaN_test["Hard_or_Paper"])['H']  # True labels (1 for hard cover, 0 for paper back)

# Features for clf
X_test_clf = ab_reduced_noNaN_test[['List Price']]  # Features used in clf

# Features for clf2
X_test_clf2 = ab_reduced_noNaN_test[['NumPages', 'Thick', 'List Price']]  # Features used in clf2

# Generate predictions
y_pred_clf = clf.predict(X_test_clf)
y_pred_clf2 = clf2.predict(X_test_clf2)
````

````python
# Confusion matrices
confusion_clf = confusion_matrix(y_test, y_pred_clf)
confusion_clf2 = confusion_matrix(y_test, y_pred_clf2)

print("Confusion Matrix for clf:\n", confusion_clf)
print("Confusion Matrix for clf2:\n", confusion_clf2)
````

The confusion matrix shows:

True Positives (TP): Correctly predicted positives (Hard Cover = 1).
True Negatives (TN): Correctly predicted negatives (Paper Back = 0).
False Positives (FP): Incorrectly predicted positives (predicted Hard Cover but was actually Paper Back).
False Negatives (FN): Incorrectly predicted negatives (predicted Paper Back but was actually Hard Cover).

````python
# Helper function to calculate metrics
def calculate_metrics(confusion):
    TP = confusion[1, 1]
    TN = confusion[0, 0]
    FP = confusion[0, 1]
    FN = confusion[1, 0]
    
    sensitivity = TP / (TP + FN) if (TP + FN) != 0 else 0  # Recall
    specificity = TN / (TN + FP) if (TN + FP) != 0 else 0
    accuracy = (TP + TN) / np.sum(confusion)
    
    return np.round([sensitivity, specificity, accuracy], 3)

# Metrics for clf
sensitivity_clf, specificity_clf, accuracy_clf = calculate_metrics(confusion_clf)

# Metrics for clf2
sensitivity_clf2, specificity_clf2, accuracy_clf2 = calculate_metrics(confusion_clf2)

print("Metrics for clf:")
print(f"Sensitivity: {sensitivity_clf}, Specificity: {specificity_clf}, Accuracy: {accuracy_clf}")

print("\nMetrics for clf2:")
print(f"Sensitivity: {sensitivity_clf2}, Specificity: {specificity_clf2}, Accuracy: {accuracy_clf2}")
````

> The way a **classification decision tree** is fit is that at each step in the construction process of adding a new **decision node splitting rule** to the current **tree structure**, all possible **decision rules** for all possible **predictor variables** are considered, and the combination that improves the **prediction** the most (as measured by the criterion of either "Gini impurity" or "Shannon entropy") and in accordance with the rules of the decision tree (such as the `max_depth` argument) is added to the **classification decision tree**. This means the relative contribution of each **predictor variable** to the overall explanatory power of the model can be calculated, and this is what the `.feature_importances_` attribute does.

But the so-called **feature importance** heuristics allows us to judge how relatively important the overall contributions from different features are in the final decision tree predictions. In linear regression, the coefficients represent the direct, linear relationship between each predictor variable and the predicted outcome. A positive coefficient indicates that as the predictor variable increases, the outcome also increases, and vice versa. Feature importances in decision trees are more about how much each feature helps separate the classes, rather than showing a direct relationship with the outcome.
